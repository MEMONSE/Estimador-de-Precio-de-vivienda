Gerero diccionario para la app

```python
import pandas as pd
# Cargar el archivo CSV
df = pd.read_csv('/Users/Usuario/Desktop/DATOS VIVOS/TP FINAL/dataset/processed.csv')

# Crear un diccionario único entre place_name y state_name
diccionario = (
    df[['place_name', 'state_name']]
    .drop_duplicates()
    .set_index('place_name')['state_name']
    .to_dict()
)

# Imprimir en formato tipo JSON (uno debajo del otro)
for place, state in diccionario.items():
    print(f'    "{place}": "{state}",')
```
